#### 1. 
For each of parts (a) through (d), indicate whether we would generally expect the performance of a flexible statistical learning method to be better or worse than an inflexible method. Justify your answer.

(a) The sample size $n$ is extremely large, and the number of predictors $p$ is small.

- An inflexible model like a linear regression would probably perform better here, because we have enough sample data to make a reasonably good prediction and the few features indicate that the model probably doesn't need so much complexity.


(b) The number of predictors $p$ is extremely large, and the number of observations $n$ is small.

- No model would perform really well with too few samples available to train. An inflexible one would probably show a hight bias. Hopefully, a flexible model would be able to catch all the nuances represented in the features. If this is possible depends on the quality of the data (Is it representative for the whole population or rather varying around it grossly?)


(c) The relationship between the predictors and response is highly non-linear.

- A flexible model performs better, or a linear model with the features transformed non-linearily before training (in case this relationship is known for some reason).


(d) The variance of the error terms, i.e. $\sigma^2 = \operatorname{Var}(\epsilon)$, is extremely high.

- A flexible model would perform worse than an inflexible one, because it would fit to well to the errors present in the training set. In the real world however, we don't see in the data, if the variance of the error term is hight or low.


- generally for (a) through (d): Measured on training data, flexible models would always perform better (with the risk of over fitting to it).

(b) Explain why each of the five curves has the shape displayed in part (a).

- (squared) bias: high bias for inflexible models as they cannot fit well to the data, decreases to 0 (on the training data) for inflexible models, because these fit tightly to the training data; convex because it is the bias squared

- variance: low for inflexible models, as their predictions will not vary so much over different test sets, increases fast for more flexible models, because these might have trained some extra bends that the test data doesn't reflect

- training error: high or medium for inflexible models and decreases for more flexible models, converges to 0

- test error: it is what it is for inflexible models, then there is a sweet spot where it decreases for the best performing models, and then increases the more flexible the model gets

- Bayes (or irreducible) error: constant (it is the same for all the models)

#### 5.
What are the advantages and disadvantages of a very flexible (versus a less flexible) approach for regression or classification? Under what circumstances might a more flexible approach be preferred to a less flexible approach? When might a less flexible approach be preferred?

- a very flexible model has the advantage of fitting very well to the present data and model very complex relationships among the features

- it's disadvantage is that it can easily over fit and then not be able to predict on unseen data

- we want to use a more flexible model when we suspect a complex relationship of how the features play into the target and we have enough training data to train on (to even out possible variation in the features)

- we want to use a simpler model if we are looking for a robust, simple solution

The table below provides a training data set containing six observations, three predictors, and one qualitative response variable.

| Obs. | X1  | X2 | X3 |  Y    |
|------|-----|----|----|-------|
| 1    |  0  |  3 |  0 | Red   |
| 2    |  2  |  0 |  0 | Red   |
| 3    |  0  |  1 |  3 | Red   |
| 4    |  0  |  1 |  2 | Green |
| 5    | −1  |  0 |  1 | Green |
| 6    |  1  |  1 |  1 | Red   |


Suppose we wish to use this data set to make a prediction for Y when X1 = X2 = X3 = 0 using K-nearest neighbors.

(a) Compute the Euclidean distance between each observation and the test point, X1 = X2 = X3 = 0.

- this stackoverflow issue describes how and why we can [use numpy.linalg.norm()](https://stackoverflow.com/questions/1401712/how-can-the-euclidean-distance-be-calculated-with-numpy) to calculate the euclidean distance: "This works because the Euclidean distance is the l2 norm, and the default value of the ord parameter in numpy.linalg.norm is 2."


````python
import numpy as np

new_observation = np.array((0,0,0))

distance_1 = np.linalg.norm(np.array((0,3,0))-new_observation)
distance_2 = np.linalg.norm(np.array((2,0,0))-new_observation)
distance_3 = np.linalg.norm(np.array((0,1,3))-new_observation)
distance_4 = np.linalg.norm(np.array((0,1,2))-new_observation)
distance_5 = np.linalg.norm(np.array((-1,0,1))-new_observation)
distance_6 = np.linalg.norm(np.array((1,1,1))-new_observation)

distances = {"distance_1":distance_1, "distance_2":distance_2, "distance_3":distance_3,
             "distance_4":distance_4, "distance_5":distance_5, "distance_6":distance_6}
````

(b) What is our prediction with K = 1? Why?

````python
import heapq

smallest_n_keys = heapq.nsmallest(1, distances, key=distances.get)
print(smallest_n_keys) 
````

- green, because the closest observation is green

(c) What is our prediction with K = 3? Why?

````python
smallest_n_keys = heapq.nsmallest(3, distances, key=distances.get)
print(smallest_n_keys) 
````

- red, because 2/3 of the closest observations are red

(d) If the Bayes decision boundary in this problem is highly non-linear, then would we expect the best value for K to be large or
small? Why?

- the best K would be small to capture local relationships rather than non-local ones
